```python

import pandas as pd
import matplotlib.pyplot as plt


```

```python

df = pd.read_csv('Stores.csv')
df.columns = df.columns.str.strip()

```

```python
sales_grouped = df.groupby('Store ID')['Store_Sales'].sum()
```

```python
plt.figure(figsize=(12, 6))
sales_grouped.plot(kind='bar', title='Total Sales by Store ID (using groupby & sum)')
plt.xlabel('Store ID')
plt.ylabel('Total Sales')
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
avg_customers = df.groupby('Store ID')['Daily_Customer_Count'].mean()
avg_customers.head()
```

```python
plt.figure(figsize=(12, 6))
avg_customers.plot(kind='bar', title='Average Daily Customer Count by Store (using groupby)')
plt.xlabel('Store ID')
plt.ylabel('Avg Daily Customers')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8, 5))
plt.scatter(df['Store_Area'], df['Store_Sales'], color='blue')
plt.title('Store Area vs Store Sales')
plt.xlabel('Store Area')
plt.ylabel('Store Sales')
plt.grid(True)
plt.tight_layout()
plt.show()

```
